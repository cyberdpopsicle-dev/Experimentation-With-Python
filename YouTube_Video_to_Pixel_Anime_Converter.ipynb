```python
import os

url = input("Enter YouTube URL: ")

os.system(f'yt-dlp -f mp4 -o "video.mp4" "{url}"')
```

```python
import cv2
import os

video_path = "video.mp4"

cap = cv2.VideoCapture(video_path)

os.makedirs("frames", exist_ok=True)

count = 0

while True:
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imwrite(f"frames/frame_{count:05d}.jpg", frame)

    count += 1

cap.release()

print("Frames Extracted:", count)
```

```python
from PIL import Image
import os
from tqdm import tqdm

os.makedirs("anime_frames", exist_ok=True)

frame_files = sorted(os.listdir("frames"))

for file in tqdm(frame_files):

    img = Image.open(f"frames/{file}")

    img = img.resize((960, 540))

    img = img.quantize(colors=128)

    img = img.resize((1280, 720), Image.NEAREST)

    # Convert to RGB before saving as JPEG
    img = img.convert('RGB')

    img.save(f"anime_frames/{file}")
```

```python
import cv2
import os

frame_folder = "anime_frames"

images = sorted(os.listdir(frame_folder))

frame = cv2.imread(os.path.join(frame_folder, images[0]))

height, width, layers = frame.shape

video = cv2.VideoWriter(
    "anime_output.mp4",
    cv2.VideoWriter_fourcc(*'mp4v'),
    30,
    (width, height)
)

for image in images:

    video.write(
        cv2.imread(os.path.join(frame_folder, image))
    )

video.release()

print("Anime Video Saved")
```
